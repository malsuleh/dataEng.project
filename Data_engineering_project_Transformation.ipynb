## **Now we work on the Dubai Dubai dataframe to replace its null values instead of dropping the rows, this is because the null values are little.**

```python
 # function to fill missing values with a random sample, we larened it in thye lab.
def impute_na(data, variable):

    df = data.copy()


    df[variable+'_random'] = df[variable]



    random_sample = df[variable].dropna().sample(df[variable].isnull().sum(), random_state=0)


    random_sample.index = df[df[variable].isnull()].index
    df.loc[df[variable].isnull(), variable+'_random'] = random_sample

    return df[variable+'_random']
```

```python
df_Dub['Actual_worth'] = impute_na(df_Dub,'actual_worth')
df_Dub = df_Dub.drop('actual_worth', axis=1)

```
